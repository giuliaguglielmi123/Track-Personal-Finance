### Custom Function

```python
def clean_df(df):
    df.columns = df.columns.str.replace(' ', '_')
    df.columns = df.columns.str.replace(' ', '')
    df.columns = df.columns.str.lower()

    return(df)
    

def decimal_converter(value):
    try:
        return float(value.replace(',', '.'))
    except ValueError:
        return value
```

### l) Upload data

```python
df_giugno = pd.read_excel("Raw/giugno2022.xlsx")
df_luglio = pd.read_excel("Raw/luglio2022.xlsx")
df_settembre = pd.read_excel("Raw/settembre2022.xlsx")
```

# clean tables and select interest columns

```python
### ... Clean dataset columns 2022
## giugno 
df_giugno_clean = clean_df(df_giugno)
df_giugno_clean = df_giugno_clean.drop(df_giugno_clean.index[23:])
df_giugno_clean["data_operazione"] = pd.to_datetime(df_giugno_clean["data_operazione"]).dt.date

## luglio 
df_luglio_clean = clean_df(df_luglio)
df_luglio_clean = df_luglio_clean.drop(df_luglio_clean.index[19:])
df_luglio_clean["data_operazione"] = pd.to_datetime(df_luglio_clean["data_operazione"]).dt.date


## settembre 
df_sett_clean = clean_df(df_settembre)
df_sett_clean = df_sett_clean.drop(df_sett_clean.index[9:])
df_sett_clean["data_operazione"] = pd.to_datetime(df_sett_clean["data_operazione"]).dt.date
```

```python
### ... Select interested columns
## giugno
df_sub_giugno = df_giugno_clean[["data_operazione", "descrizione_operazione", "importo_in_euro"]]
### ... Add Year and Month column
df_sub_giugno["year"] = pd.DatetimeIndex(df_sub_giugno["data_operazione"]).year
df_sub_giugno["month"] = pd.DatetimeIndex(df_sub_giugno["data_operazione"]).month

## luglio
df_sub_luglio = df_luglio_clean[["data_operazione", "descrizione_operazione", "importo_in_euro"]]
### ... Add Year and Month column
df_sub_luglio["year"] = pd.DatetimeIndex(df_sub_luglio["data_operazione"]).year
df_sub_luglio["month"] = pd.DatetimeIndex(df_sub_luglio["data_operazione"]).month

## settembre
df_sub_sett = df_sett_clean[["data_operazione", "descrizione_operazione", "importo_in_euro"]]
### ... Add Year and Month column
df_sub_sett["year"] = pd.DatetimeIndex(df_sub_sett["data_operazione"]).year
df_sub_sett["month"] = pd.DatetimeIndex(df_sub_sett["data_operazione"]).month
```

```python
## There are some values with point as decimal separator, need to replace them !
df_sub_giugno["importo_in_euro"] = df_sub_giugno["importo_in_euro"].astype(str).str.replace(",", ".")
```

```python
dataset_finance = df_sub_giugno.append(df_sub_luglio)
dataset_finance = dataset_finance.append(df_sub_sett)
dataset_finance["importo_in_euro"] = dataset_finance["importo_in_euro"].astype(float)
dataset_finance
```

```python
## Remove digits from column - Lower case
## giugno
dataset_finance["descrizione_operazione"] = dataset_finance["descrizione_operazione"].str.lower()
dataset_finance["descrizione_operazione"] = dataset_finance["descrizione_operazione"].str.replace('\d+|eur|,', '')
```

### ll) Categories Column

```python
# create a dictionary of mappings
mapping = {"ristorante": "pranzo o cena fuori", 
           "pizza": "pranzo o cena fuori",
           "bistrot": "pranzo o cena fuori",
           "farmacia": "farmacia",
           "farmacie": "farmacia",
           "supermercato": "spesa supermercato",
           "supermarket": "spesa supermercato",
           "supermercati": "spesa supermercato",
           "market": "spesa supermercato",
           "italki": "lezioni di inglese",
           "verbling": "lezioni di inglese",
           "amazon": "amazon",
           "amzn": "amazon",
           "edreams": "voli",
           "caffetteria": "bar",
           "ryanair": "voli",
           "bar": "bar",
           "prelievo": "prelievo",
           "prelievi": "prelievo",
           "telecomital": "ricarica telefono"}
```

```python
def map_column(x):
    for key in mapping:
        if key.lower() in x.lower():
            return mapping[key]
    return "altro"

dataset_finance["category"] = dataset_finance["descrizione_operazione"].map(lambda x: map_column(x))

# dataset_money.loc[dataset_money["category"] == "Altro"]
```
